```python
import requests
from bs4 import BeautifulSoup
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# 1. R√©cup√©rer les donn√©es depuis Wikipedia
url = "https://fr.wikipedia.org/wiki/D%C3%A9mographie_du_Canada"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

tables = soup.find_all("table", {"class": "wikitable"})
target_table = next(
    (
        table
        for table in tables
        if all(
            col in [th.get_text(strip=True) for th in table.find_all("th")]
            for col in ["Ann√©e", "Population", "√âvolution"]
        )
    ),
    None,
)

data = []
if target_table:
    for row in target_table.find_all("tr")[1:]:
        cols = row.find_all(["td", "th"])
        if len(cols) >= 3:
            year = cols[0].get_text(strip=True)
            pop_str = cols[1].get_text(strip=True).replace("\xa0", "").replace(" ", "")
            try:
                population = int(pop_str)
            except ValueError:
                population = None

            if population is not None:
                data.append((int(year), population))
else:
    print("Tableau cible non trouv√©")
    exit(1)

# 2. Pr√©parer les donn√©es pour le mod√®le
years = np.array([item[0] for item in data]).reshape(-1, 1)
populations = np.array([item[1] for item in data])

# 3. Entra√Æner un mod√®le de r√©gression lin√©aire
model = LinearRegression()
model.fit(years, populations)

# 4. Faire une pr√©diction pour 2025
year_to_predict = np.array([[2025]])
predicted_population = model.predict(year_to_predict)[0]

print(f"Population pr√©dite pour 2025 : {int(predicted_population):,}".replace(",", " "))

# 5. Visualisation avec matplotlib
plt.figure(figsize=(8,5))

# Courbe des donn√©es historiques
plt.plot(years, populations, marker='o', label="Population historique")

# Point de la pr√©diction 2025
plt.scatter(year_to_predict, predicted_population, color='red', label="Pr√©diction 2025", zorder=5)

# Optionnel : tracer la droite de r√©gression
years_all = np.arange(years.min(), 2026).reshape(-1,1)
pop_pred_line = model.predict(years_all)
plt.plot(years_all, pop_pred_line, linestyle='--', color='gray', label="R√©gression lin√©aire")

plt.title("Population du Canada : Historique et pr√©diction 2025")
plt.xlabel("Ann√©e")
plt.ylabel("Population")
plt.xticks(np.arange(years.min(), 2026, 1), rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
